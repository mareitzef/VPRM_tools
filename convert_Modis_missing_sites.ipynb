```python
import pandas as pd
import os

# downloaded data manually as MODIStool did not work for the sites:
    # RU-Fyo,ENF,56.4615,32.9221
    # IT-Col,DBF,41.8494,13.5881
    # FI-Sod,ENF,67.3619,26.6378
    # DE-Tha,ENF,50.9636,13.5669
    # NL-Loo,ENF,52.1666,5.7436
    # DE-Spw,WET,51.8923,14.0337
# and to check if it is the same data with Modis Tool:
    # AT-Neu,GRA,47.1167,11.3175


# https://appeears.earthdatacloud.nasa.gov/
path = "/home/madse/Downloads/missing_sties/"
# terra satellite
MOD09A1 = pd.read_csv(path+"MOD09A1-061-results.csv")
MOD13Q1 = pd.read_csv(path+"MOD13Q1-061-results.csv")
MOD15A2H = pd.read_csv(path+"MOD15A2H-061-results.csv")
# aqua satellite
MYD09A1 = pd.read_csv(path+"MYD09A1-061-results.csv")
MYD13Q1 = pd.read_csv(path+"MYD13Q1-061-results.csv")
MYD15A2H = pd.read_csv(path+"MYD15A2H-061-results.csv")


import os

# Group 1: MOD15A2H and MYD15A2H
aqua_terra_str = ["MOD15A2H", "MYD15A2H"]
bands = ["Fpar_500m", "Lai_500m"]

for i, df in enumerate([MOD15A2H, MYD15A2H]):
    for band in bands:
        # Group by ID and extract desired columns
        band_df = df.groupby("ID")[["Date", f"{aqua_terra_str[i]}_061_{band}"]].apply(lambda x: x.reset_index(drop=True))

        # Rename columns
        band_df.rename(columns={"Date": "calendar_date", f"{aqua_terra_str[i]}_061_{band}": "value"}, inplace=True)

        # Define output directory
        output_directory = os.path.join(path, "output")
        os.makedirs(output_directory, exist_ok=True)

        # Write to separate Excel files
        for group_name, group_data in band_df.groupby(level=0):
            filename = os.path.join(output_directory, f"{group_name}_{aqua_terra_str[i]}_{band}_2000-02-18_2015-01-17.xlsx")
            group_data.reset_index(drop=True, inplace=True)  # Drop index just before writing
            group_data.to_excel(filename, index=False)
            print(f"File '{filename}' written.")

# Group 2: MOD09A1 and MYD09A1
aqua_terra_str_2 = ["MOD09A1", "MYD09A1"]
bands_2 = ["sur_refl_b02", "sur_refl_b06"]

for i, df in enumerate([MOD09A1, MYD09A1]):
    for band in bands_2:
        # Group by ID and extract desired columns
        band_df = df.groupby("ID")[["Date", f"{aqua_terra_str_2[i]}_061_{band}"]].apply(lambda x: x.reset_index(drop=True))

        # Rename columns
        band_df.rename(columns={"Date": "calendar_date", f"{aqua_terra_str_2[i]}_061_{band}": "value"}, inplace=True)

        # Define output directory
        output_directory = os.path.join(path, "output")
        os.makedirs(output_directory, exist_ok=True)

        # Write to separate Excel files
        for group_name, group_data in band_df.groupby(level=0):
            filename = os.path.join(output_directory, f"{group_name}_{aqua_terra_str_2[i]}_{band}_2000-02-18_2015-01-17.xlsx")
            group_data.reset_index(drop=True, inplace=True)  # Drop index just before writing
            group_data.to_excel(filename, index=False)
            print(f"File '{filename}' written.")


# Group 3: MOD13Q1 and MYD13Q1
aqua_terra_str_3 = ["MOD13Q1", "MYD13Q1"]
bands_3 = ["250m_16_days_EVI"]

for i, df in enumerate([MOD13Q1, MYD13Q1]):
    for band in bands_3:
        # Group by ID and extract desired columns
        band_df = df.groupby("ID")[["Date", f"{aqua_terra_str_3[i]}_061__{band}"]].apply(lambda x: x.reset_index(drop=True))

        # Rename columns
        band_df.rename(columns={"Date": "calendar_date", f"{aqua_terra_str_3[i]}_061__{band}": "value"}, inplace=True)

        # Define output directory
        output_directory = os.path.join(path, "output")
        os.makedirs(output_directory, exist_ok=True)

        # Write to separate Excel files
        for group_name, group_data in band_df.groupby(level=0):
            filename = os.path.join(output_directory, f"{group_name}_{aqua_terra_str_3[i]}_{band}_2000-02-18_2015-01-17.xlsx")
            group_data.reset_index(drop=True, inplace=True)  # Drop index just before writing
            group_data.to_excel(filename, index=False)
            print(f"File '{filename}' written.")
```
